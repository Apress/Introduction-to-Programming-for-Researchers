```python
# script by sanjiv r h-k
%matplotlib qt
from mpl_toolkits import mplot3d
import pandas as pd
import matplotlib.pyplot as plt
import os; home = os.path.expanduser('~')
# we need the starTable's label line
starTable = pd.read_csv(home + '/datafile/hygdata_v41.csv')
label_line = starTable.columns.tolist()
# now we label the columns in the big dipper dataframe
big_dipper = pd.read_csv(home + '/datafile/big_dipper.txt',\
index_col=0,\
names=label_line)

# Background color
bkgrd_color = '#050540'
# make 3D rotatable plot of big dipper
fig = plt.figure(facecolor= bkgrd_color)  # Set the figure background to black
ax = fig.add_subplot(111, projection='3d', facecolor= bkgrd_color)  # Set 3D axes background to black
x, y, z = big_dipper['x'], big_dipper['y'], big_dipper['z']
ax.scatter(x, y, z, c = 'azure', depthshade = False)
# Show names of stars - code snippet provided by ChatGPT
# for xi, yi, zi, name in zip(x, y, z, big_dipper['proper']):
#     ax.text(xi, yi, zi, name, color='white', fontsize=10)

# ChatGPT provided code
# Make axes lines invisible
ax.xaxis.line.set_color(bkgrd_color)  # X axis line
ax.yaxis.line.set_color(bkgrd_color)  # Y axis line
ax.zaxis.line.set_color(bkgrd_color)  # Z axis line

# Make tick marks invisible
ax.tick_params(axis='x', colors= bkgrd_color)
ax.tick_params(axis='y', colors= bkgrd_color)
ax.tick_params(axis='z', colors= bkgrd_color)

# Source - https://stackoverflow.com/a/50435041
# Posted by S.A.
# Retrieved 2026-04-02, License - CC BY-SA 4.0

# Get rid of colored axes planes
# First remove fill
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False

# Now set color to white (or whatever is "invisible")
ax.xaxis.pane.set_edgecolor('#050540')
ax.yaxis.pane.set_edgecolor('#050540')
ax.zaxis.pane.set_edgecolor('#050540')

# Bonus: To get rid of the grid as well:
ax.grid(False)
ax.view_init(elev= -30, azim= 0, roll = -9)  # Adjust these numbers to match Earth's view
plt.show()
```
